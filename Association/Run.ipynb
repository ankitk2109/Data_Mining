## Q1

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
```

```python
gpa_df = pd.read_csv(r"D:\Study\DM_Practicals\Practical_3\practical_3_ASSOCIATION\specs\gpa_question1.csv")
```

```python
gpa_df.head()
```

```python
gpa_df = gpa_df.drop('count', axis=1)
```

```python
gpa_df
```

```python
cat_df = pd.get_dummies(gpa_df, prefix = 'category')
```

```python
cat_df
```

```python
freq_itemsets = apriori(cat_df, min_support=0.15 ,use_colnames=True)
```

```python
freq_itemsets
```

```python
export = freq_itemsets.to_csv(r'D:\Study\DM_Practicals\Practical_3\practical_3_ASSOCIATION\output\question1_out_apriori.csv',index=False)
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
ass_rule_9 = association_rules(freq_itemsets, metric="confidence", min_threshold=0.9)
ass_rule_9
```

```python
export = ass_rule_9.to_csv(r'D:\Study\DM_Practicals\Practical_3\practical_3_ASSOCIATION\output\question1_out_rules9.csv',index=False)
```

```python
ass_rule_7 = association_rules(freq_itemsets, metric="confidence", min_threshold=0.7)
ass_rule_7
```

```python
export = ass_rule_7.to_csv(r'D:\Study\DM_Practicals\Practical_3\practical_3_ASSOCIATION\output\question1_out_rules7.csv',index=False)
```

## Q2 FP Growth

```python
import pandas as pd
file= pd.read_csv(r"D:\Study\DM_Practicals\Practical_3\practical_3_ASSOCIATION\specs\bank_data_question2.csv")
file
```

```python
df= file.drop('id',axis=1)
df
```

```python
df['age'] = pd.cut(df['age'],3)
df['income'] = pd.cut(df['income'],3)
df['children'] = pd.cut(df['children'],3)
```

```python
df
```

```python
category_df = pd.get_dummies(df)
```

```python
category_df
```

```python
from mlxtend.frequent_patterns import fpgrowth
frequent_sets = fpgrowth(category_df, min_support=0.2,use_colnames=True)
frequent_sets
```

```python
from mlxtend.frequent_patterns import association_rules
ass_rule10= association_rules(frequent_sets, metric="confidence", min_threshold=0.8)
ass_rule10
```

```python
ass_rule10.to_csv(r'D:\Study\DM_Practicals\Practical_3\practical_3_ASSOCIATION\output\question2_out_rules.csv',index=False)
```
